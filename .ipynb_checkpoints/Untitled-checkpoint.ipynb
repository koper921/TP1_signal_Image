```python
import numpy as np
f_grey= np.zeros(f.shape)

for i in range(f_grey.shape[0]):
    for j in range(f_grey.shape[1]):
        if f[i,j,0]> 200 or f[i,j,1]> 200 or f[i,j,2]> 200: # ranjouter and < 250
            f_grey[i,j,0]= 1#f[:,:,0]*299/1000
            f_grey[i,j,1]= 1  #f[:,:,0]*587/1000
            f_grey[i,j,2]= 1  
        else: 
            f_grey[i,j,0]= 0#f[:,:,0]*299/1000
            f_grey[i,j,1]= 0  #f[:,:,0]*587/1000
            f_grey[i,j,2]= 0  #f[:,:,0]*114/1000

plt.imshow(f_grey)
plt.show()

```

```python
f_quantification = np.zeros(f.shape)

for i in range(f_quantification.shape[0]):
    for j in range(f_quantification.shape[1]):
        
        for k in range(f_quantification.shape[2]):

            if f[i,j,k]< 64 and f[i,j,k]>=0 : # ranjouter and < 250
                f_quantification[i,j,k]= 0#f[:,:,0]*299/1000
                  
            if f[i,j,k]< 128 and f[i,j,k]>=64:
                f_quantification[i,j,k]= 1
                
            if f[i,j,k]< 192 and f[i,j,k]>=128:
                f_quantification[i,j,k]= 2
            
            if f[i,j,k]>=192:
                f_quantification[i,j,k]= 3
               
                
plt.imshow(f_quantification)
plt.show()
```

```python
import numpy as np
import math
 
def carre (x):
    return x*x

def f_gauss(z, sigma) :
    return np.exp((-np.power(z,2)/2*carre(sigma)))

def bordure_image(image, bord):
    image2= np.zeros((image.shape[0]+2*bord, image.shape[1]+2*bord, image.shape[2] ))
    image2[bord:,bord:,:]= image
    return image2
    

def u_noised(p1, p2, w, u,sigma_i, sigma_s):
    #vois = np.zeros(u.shape)
    u_noised =0
    constante = 0
    for i in range(0,2*w+1):
        for j in range(0,2*w+1):
            u_noised = u_noised + np.exp(-(carre(i-w-1)+carre(j-w-1))/2*carre(sigma_s)) * u[i+p1-w-1, j+p2-w-1] * f_gauss(carre(u[i+p1-w-1, j+p2-w-1]-u[w+1+p1-w-1, w+1+p2-w-1] ), sigma_i)
            constante = constante + np.exp(-(carre(i-w-1)+carre(j-w-1))/2*carre(sigma_s))* f_gauss(carre(u[i+p1-w-1, j+p2-w-1]-u[w+1+p1-w-1, w+1+p2-w-1] ), sigma_i)
   # print(constante)
    #print(u_noised )
    #print(u[p1,p2])
    return u_noised * 1/constante   

  
    

f2 =f[:,:,2]  
print(f2.shape )

print(u_noised(44, 99, 8, f2, 0.5,0.4))    
        
plt.imshow(f2)
plt.show()

      
            
    
```
